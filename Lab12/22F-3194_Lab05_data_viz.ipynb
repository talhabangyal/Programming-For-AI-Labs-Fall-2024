# Assignment 3: Subplots

Filter the transactions table to stores 1, 2, 3 and 4, then reshape it so each row is a date, and the columns of the table are the transactions for each store. 

Once you've done that, create a 2x2 line plot of transactions vs time, with one plot for each store.

Thanks!

```python
transactions = pd.read_csv("../Lab12/retail/transactions.csv")
```

```python
transaction = transactions[transactions['store_nbr'].isin([1, 2, 3, 4])]
table = transaction.pivot_table(index='date', columns='store_nbr', values='transactions')
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
store_1234 = [1, 2, 3, 4]
for i, store in enumerate(store_1234):
    ax = axes[i//2, i%2]
    ax.plot(table.index, table[store], color='red', label=f'Store {store}')
    ax.set_title(f'Store {store}')
    ax.set_xlabel('Date')
    ax.set_ylabel('Transactions')
    ax.legend()
plt.tight_layout()
plt.show()
```
